```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the iris dataset
df = pd.read_csv('iris.csv')

# Function to plot histograms for each column
def plotPerColumnDistribution(df, n_rows, n_cols):
    """Plots histograms for each column in a DataFrame.

    Args:
        df: The DataFrame to plot.
        n_rows: The number of rows in the subplot grid.
        n_cols: The number of columns in the subplot grid.
    """
    fig, axes = plt.subplots(nrows=n_rows, ncols=n_cols, figsize=(16, 8))

    # Flatten the axes array for easier iteration
    axes = axes.flatten()

    # Loop through each column and plot a histogram
    for i, col in enumerate(df.columns):
        axes[i].hist(df[col], bins=10)
        axes[i].set_title(f"{col} (column {i})")
        axes[i].set_xlabel("Value")
        axes[i].set_ylabel("Counts")
        axes[i].grid(True)

    # Remove any unused subplots
    plt.tight_layout()
    plt.show()

# Call the function to generate the histograms
plotPerColumnDistribution(df, 3, 2)
```

```python

import matplotlib.pyplot as plt

# Assuming 'df' is your DataFrame containing 'SepalLengthCm' and 'SepalWidthCm' columns

plt.figure()
plt.scatter(df['SepalLengthCm'], df['SepalWidthCm'], label='SepalLengthCm vs SepalWidthCm', color='blue')
plt.xlabel('SepalLengthCm')
plt.ylabel('SepalWidthCm')
plt.title('SepalLengthCm vs SepalWidthCm')
plt.legend()
plt.grid(True)  # Adding grid lines for better readability
plt.show()
```

```python
import matplotlib.pyplot as plt

# Assuming 'df' is your DataFrame containing 'SepalLengthCm' and 'SepalWidthCm' columns

# Creating a condition to differentiate points based on a threshold (example)
threshold = df['SepalLengthCm'].mean()

plt.figure()
plt.scatter(df[df['SepalLengthCm'] >= threshold]['SepalLengthCm'], 
            df[df['SepalLengthCm'] >= threshold]['SepalWidthCm'], 
            label='SepalLengthCm >= Threshold', color='blue')
plt.scatter(df[df['SepalLengthCm'] < threshold]['SepalLengthCm'], 
            df[df['SepalLengthCm'] < threshold]['SepalWidthCm'], 
            label='SepalLengthCm < Threshold', color='red')
plt.xlabel('SepalLengthCm')
plt.ylabel('SepalWidthCm')
plt.title('SepalLengthCm vs SepalWidthCm')
plt.legend()
plt.grid(True)  # Adding grid lines for better readability
plt.show()
```
